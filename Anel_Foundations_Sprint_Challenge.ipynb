```
x = [[0,1,2,3,4],[-6.28,-3.14,0,3.14,6.28],[0,0,0,0,0]]

deltax0 = float(x[1][1]) - float(x[1][0])

m = deltax0
x = np.array(x, dtype=float)
m  = (x[1][1] - x[1][0])/(x[0][1] - x[0][0])
b = x[1][0] - (m*x[0][0])


def f(x):
  return m*x+b

plt.plot(x[0], x[1],'g^')

plt.plot(x[0], f(x[0]), 'r--')


plt.text(0.6, 0.6, r'$\mathrm{f}(x) = 3.14x-6.28$',
         fontsize=20)


# x[2] - is constant 
```

### 9. invert a matrix and describe what it does

$MM^{-1} = I$

$M = \begin{pmatrix} 1.1 & 0.9 \\ 0.8 & 1.2 \end{pmatrix}$

```
m = [[1.1, 0.9],[0.8, 1.2]]

print(m)

mi=np.linalg.inv(m)

print(mi)

# MM^{-1}
```
